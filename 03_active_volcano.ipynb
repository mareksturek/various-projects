```python
# Active volcano

# Import dataset
os.chdir('//Users/mareksturek/Documents/GitHub/various-projects/data')
dataset = pd.read_csv('volcano.csv')
```

```python
# Show on map in red

lat_vo = list(dataset['Latitude'])
lon_vo = list(dataset['Longitude'])
name_vo = list(dataset['Name'])

vol = fo.FeatureGroup(name='Map of World')

for lat,lon,name in zip(lat_vo,lon_vo,name_vo):
    vol.add_child(fo.Marker(location=[lat,lon] ,
                          popup=name , icon=fo.Icon(color='red')))
    
map.add_child(vol)    
```
